```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings 
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.keys import Keys
import requests
import re
```

```python
driver=webdriver.Chrome()
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
rank=[]
```

```python
name=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
try:
    for i in name:
        rank.append(i.text[:-1])
except NoSuchElementException:
    rank.append('N/A')


```

```python
v_name=[]
artist=[]
views=[]
date=[]
```

```python

name=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
try:
    for i in name:
        v_name.append(i.text[1:-5])
except NoSuchElementException:
    v_name.append('N/A')

    
try:
    name1=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in name1:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('N/A')


try:
    name2=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in name2:
        views.append(i.text)
except NoSuchElementException:
    views.append('N/A')


try:
    date1=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in date1:
        date.append(i.text)
except NoSuchElementException:
    date.append('N/A')
```

```python
import pandas as pd
df=pd.DataFrame()
df["Rank"]=rank
df["Name"]=v_name
df["Artist"]=artist
df["Views"]=views
df["Date"]=date
```
