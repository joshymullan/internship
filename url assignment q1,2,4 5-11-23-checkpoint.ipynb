```python
prod_url=[]
for page in range(0,1):
    url1=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
    for t in url1:
        prod_url.append(t.get_attribute('href'))
  #  next_button=driver.find_elements(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[26]/div/div/nav/a[11]/span')
```

```python
prod_url
```

```python
Brand_Name=[]
Smartphone_name=[]
Colour=[]
RAM=[]
Storage=[]
Primary_Camera=[] 
Secondary_Camera=[]
Display_Size=[]
Battery_Capacity=[]
Price=[]
Product_URL=[]
```

```python
for i in prod_url:
    driver.get(i)
    time.sleep(1)
   
    try:
        brand=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')  
        Brand_Name.append(brand.text)
        
        price=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[4]/div[1]/div/div[1]')
        Price.append(price.text)
        
        storage=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        Storage.append(storage.text)
        
        colour=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        Colour.append(colour.text)
        
        display=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]')
        Display_Size.append(display.text)
        
        pricam=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]')
        Primary_Camera.append(pricam.text)
        
        battery=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]')
        Battery_Capacity.append(battery.text)
        
        product=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[3]/div/div[2]/div[1]/p')
        Product_URL.append(product.text)
    except NoSuchElementException :
        Brand_Name.append('-')
        Price.append('-')
        Storage.append('-')
        Colour.append('-')
        Display_Size.append('-')
        Primary_Camera.append('-')
        Battery_Capacity.append('-')
        Product_URL.append('-')
        
 
```

```python
Brand_Name,Price,Storage,Colour,Display_Size,Primary_Camera,Battery_Capacity,Product_URL
```

```python
import pandas as pd
FlipKart=pd.DataFrame([])
FlipKart['Brand_Name']=Brand_Name
FlipKart['Storage_RAM_ROM']=Storage
FlipKart['Amount P_F_Camera']=Primary_Camera
FlipKart['Display_size']=Display_Size
FlipKart['Battery']=Battery_Capacity
FlipKart['Price']=Price

FlipKart
```
